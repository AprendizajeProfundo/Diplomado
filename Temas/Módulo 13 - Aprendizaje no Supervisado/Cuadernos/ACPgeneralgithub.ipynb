# Introducción
En este documento se llama análisis en componentes principales (ACP) canónico, descrito en el cuaderno [Análisis en componentes principales](https://github.com/AprendizajeProfundo/Diplomado/blob/master/Temas/M%C3%B3dulo%2014-%20Aprendizaje%20no%20Supervisado/Cuadernos/ACP.ipynb), que corresponde al que se apoya en la geometría eucliana también canónica (la usual). En este cuaderno se hace una extensión que sirve para enmarcar estos métodos, denominados factoriales en un marco de referencia común, dentro de los cuales están lo métodos de análisis de correspondencias simples y múltiples. La presentación completa se presenta en el capítulo cuarto del libro [Estadística descriptiva multivariada](https://bit.ly/3mD9rEY). La extensión se denomina ACP generalizado o general. El ACP canónico de puede escribir como el $ACP(\mathbf{X,I_p,\frac{1}{n} \mathbf{I}_n})$, donde **X** es la matriz de datos centrados o estandarizados, con $n$ filas y $p$ columnas. $\mathbf{I}_p$ es la matriz identidad $p\times p$ y $\frac{1}{n} \mathbf{I}_n$ la matriz didentidad $n x n$. En la  generalización se introducen otras matrices **M** y **N** y tal ACP se escribe *ACP(**X,M,N**)*.

# Geometría euclidiana multidimensional generalizada

## Métrica

 **M**  matriz que define un producto interno en un espacio vectorial en los reales, $E$.
 A partir de **M** se dota al espacio de una geometría Euclidiana.
 Sean **x** y **y** dos vectores en $E$, entonces:

**M**-producto punto:  $<\mathbf{x},\mathbf{y} >_{\mathbf{M}}= \mathbf{x'My} =\mathbf{y'Mx} $
 
**M**-norma:  $\| \mathbf{x} \|_{\mathbf{M}} = \sqrt{<\mathbf{x},\mathbf{x} >_{\mathbf{M}}}$

**M**-distancia:  $d_{\mathbf{M}}(\mathbf{x},\mathbf{y}) = \| \mathbf{x} - \mathbf{y} \|_{\mathbf{M}} $

Vector **M**-unitario: $\| \mathbf{u} \|_{\mathbf{M}} = 1 \implies   \mathbf{u'Mu} = 1$
    
**M**-proyección sobre u: $<\mathbf{x},\mathbf{u} >_{\mathbf{M}}= \mathbf{x'Mu}  $
    
M-coseno:] $Cos_{\mathbf{M}} (\mathbf{x} ,\mathbf{y} ) = \dfrac{<\mathbf{x},\mathbf{y} >_{\mathbf{M}}}{\| \mathbf{x} \|_{\mathbf{M}} \| \mathbf{y} \|_{\mathbf{M}} }$
    
M-ortogonalidad: $ <\mathbf{x},\mathbf{y} >_{\mathbf{M}} = \mathbf{x'My}  = 0$

## Diagrama de dualidad ("incompleto")

Como no se dispone de la matriz de datos no se puede conectar la parte superior con la parte inferior. Las coordenadas de las variables se obtienen de los vectores propios de la matriz de varianzas y covarianzas o de la matriz de correlaciones (que es la matriz de covarianzas de los datos estandarizados). De las fórmulas del ACP general, la coordenada de una variable sobre un eje $s$ es:
$$
G_s = \sqrt{\lambda_s} \mathbf{u}_s
$$

![ACP canónico en ade4](https://raw.githubusercontent.com/AprendizajeProfundo/Diplomado/master/Temas/Módulo%2014-%20Aprendizaje%20no%20Supervisado/Imagenes/DuDiV.png)

### Matriz de varianzas y covarianzas de los resultados en las componentes del examen examen de admisión

```r
# Ejemplo matriz de covarianzas de los  resultado del examen
library(FactoClass)
data(admi)
""
"Matriz de covarianza de las cinco componentes del examen de admisión"
V<-cov(admi[,2:6])
round(V,3)
```

### Imagen euclidiana de la matriz de la varianzas y covarianzas de los resultados de las componentes en el examen de admisión 

```r
V # matriz de varianzas. y covarianzas
eigV <- eigen(V)
Lambda<-diag(eigV$values)
U<-eigV$vectors
G<-U%*%sqrt(Lambda)
rownames(G)<-colnames(V)
library(ade4)
s.arrow(G)
```

Las longitudes de los vectores en $\mathbb{R}^5$ son las desviaciones estándar, en la proyección en el primer plano factorial se acortan por la proyección unos más que otros. Eventualmente alguno puede conservar su longitud si queda sobre el plano de proyección.

```r
# desviaciones estándar
t(round(sqrt(diag(V)),2))
```

### Matriz de correlaciones de los resultados de las componentes del examen de admisión

```r
# matriz de correlaciones
V <- cor(admi[,2:6])
round(as.dist(V),3)
```

### Imagen euclidiana de la matriz de correlaciones de los resultados de las componentes del examen de admisión

```r
eigV <- eigen(V)
Lambda<-diag(eigV$values)
U<-eigV$vectors
G<-U%*%sqrt(Lambda)
rownames(G)<-colnames(V)
library(ade4)
s.corcircle(G,fullcircle = F)
```

En este caso, de la matriz de correlaciones, en  $\mathbb{R}^5$ las longitudes de los vectores variables son 1 para todas. Lo que interesa es el ángulo entre ellos, relacionado también con su distancia. Además, entre más se acerquen a la circunferencia de radio 1, están mejor representados o proyectados. No se deben interpretar en un plano variables que estén muy cortas, es posible que algunas estén mejor representada en otros planos. En el caso de esta matriz, se puede considerar que las variables están bien representadas y por lo tanto es válido interpretar todas las correlaciones. Por un lado, los resultados de sociales y textual están correlacionados, lo mismo que matemáticas y ciencias; los de imagen practicamente no están correlacionados con los de sociales y textual y algo correlacionados con matemáticas y ciencias. Nótese que éste círculo de correlaciones, ayuda a la lectura de la matriz de correlaciones y establecer grupos de variables, sobre todo cuando se tengan muchas variables.

```r
# Matriz de correlaciones
as.dist(round(V,2))
```

# Análisis en coordenadas principales (ACO)

Cuando se tienen matrices de distancias euclidianas entre todos los individuos de un estudio también se pueden obtener imágenes euclidianas de estas matrices. El procedimiento es obtener primero, la matriz de productos escalares entre individuos **W**, luego encontrar la matriz **WN** y hallar sus valores y vectores propios: {$\lambda_s, \ \mathbf{v}_s\ : \ \ s=1,2 \cdots S$}, . Las coordenadas de los individuos sobre un eje se obtienen mediante:
$$
F_s = \sqrt{\lambda_s} \mathbf{v}_s
$$

## Obtención de las matriz W a partir la matriz de distancias D y los pesos $\mathbf{N} = diag(p_i)$

\begin{equation}
w_{il}=\frac{1}{2}( d^2_{i\cdot} + d^2_{l\cdot}  -
d^2_{il} - d^2_{\cdot\cdot})
\end{equation}
donde:
$$ d^2_{i\cdot} = \sum\limits_{l=1}^n p_i d^2_{il}\ \ \ \text{y} \ \ \
d^2_{\cdot\cdot} = \sum\limits_{i,l=1}^n p_i p_l d^2_{il}$$

## Ejemplo imagen euclidiana de la matriz de distancias el ejemplo café con todas las variables, estandarizadas

### La tabla de datos

```r
data(cafe)
cafe
```

### Obtención de la matriz de distancias

```r
D <- dist(scale(cafe))
round(D,1)
```

### Imagen euclidiana de la matriz de distancias

```r
aco <- dudi.pco(D,scannf=FALSE)
aco
```

#### Selección del subespacio de representación 

```r
# Valores propios
round(inertia(aco)$tot.inertia,2)->ineraco
ineraco
"Promedio de valores propios"
round(mean(ineraco$inertia),2)
```

Tres valores propios están por encima del promedio y acumulan el 88.7\% de la inercia.

```r
# Diagrama de barras de los valores propios
# poner clase pca en aco para que fviz_eig de factoextra se ejecute
attributes(aco)$class<-c("pca","dudi") 
library(factoextra)
fviz_eig(aco)
```

El criterio de la forma del "histograma" de valores propios también sugiere usar 3 ejes
para el subespacio de representación. 

```r
# para agregar el tercer eje a las tablas del objeto aco
aco<-redo.dudi(aco,newnf=3)
```

```r
# primer plano factorial
plot(aco,gg=T,Tcol=FALSE)
```

Se observa que el primer eje ordena los cafes de más contaminados (a la izquierda del plano) y el segundo eje separa los de tostación clara (abajo) de los de tostación oscura (arriba).

```r
# plano factorial 1-3
plot(aco,1,3,gg=T,Tcol=FALSE)
```

Nótese que el tercer eje separa los dos cafés comerciales de los 10 tratamientos experimentales.
